```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
pip install boto3
```

```python
 


# Kafka Consumer Setup
consumer = KafkaConsumer(
    'demo_testing2',
    # for local machine
    bootstrap_servers=['localhost:9092'],  # add your IP here 
    # bootstrap_servers=['{add your EC2 instance public ip here}:9092']  #for ec2
    value_deserializer=lambda x: loads(x.decode('utf-8')))

 
 

 
```

```python
s3 = boto3.client('s3', aws_access_key_id='your id', aws_secret_access_key='your secret key',region_name='your region')
 
```

```python
for count, message in enumerate(consumer):
    # Define the S3 key (object key)
    s3_key = "stock_market_{}.json".format(count)

    # Convert the Kafka message value to JSON
    json_data = message.value

    # Upload the JSON data to S3
    s3.put_object(
        Bucket='kafka-stock-market-video-mudassir',  # Replace with your S3 bucket name
        Key=s3_key,
        Body=json.dumps(json_data)  # Convert JSON data to string
    )
    print("Uploaded:", s3_key) 
```
